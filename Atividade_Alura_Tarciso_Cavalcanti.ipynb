# **COMANDOS INICIAIS**

```python
import pandas as pd
url = "https://gist.githubusercontent.com/tgcsantos/3bdb29eba6ce391e90df2b72205ba891/raw/22fa920e80c9fa209a9fccc8b52d74cc95d1599b/dados_imoveis.csv"
dados = pd.read_csv(url); dados
```

## **1 - Realizar a média da metragem para cada um dos bairros**

```python
dados.groupby('Bairro').mean()["Metragem"]
```

## **2 - 2 formas de relacionar os dados por bairro (Consultar os métodos na documentação do pandas)**

```python
# 1ª Forma:
dados.groupby('Bairro').mean()
```

```python
# 2ª Forma
dados.sort_values(by='Bairro')
```

## **3 - Explorar alguns gráficos na documentação e aplicar nas demais colunas do DF, tentar colocar alguma conclusão**

```python
import numpy as np
import matplotlib.pyplot as plt

grafico_bairro = dados["Bairro"].value_counts()
grafico_metragem = dados["Metragem"].value_counts()
grafico_quartos = dados["Quartos"].value_counts()
grafico_banheiros = dados["Banheiros"].value_counts()
grafico_vagas = dados["Vagas"].value_counts()
grafico_valor = dados["Valor"].value_counts()


plt.figure(figsize = ((25, 20)))

plt.subplot(4, 2, 1)
grafico_bairro.head(10).plot.bar()
plt.title("Bairro - 10 Primeiras Colunas", fontsize = 16)

plt.subplot(4, 2, 2)
grafico_bairro.tail(10).plot.bar()
plt.title("Bairro - 10 Últimas Colunas", fontsize = 16)

plt.subplot(4, 2, 3)
grafico_bairro.plot.line()
plt.title("Bairro em Linha", fontsize = 16)

plt.subplot(4, 2, 4)
grafico_metragem.plot.hist()
plt.title("Histograma de Metragem", fontsize = 16)

plt.subplot(4, 2, 5)
grafico_quartos.plot.bar()
plt.title("Quartos em Barras", fontsize = 16)

plt.subplot(4, 2, 6)
grafico_banheiros.plot.bar()
plt.title("Banheiros em Barras", fontsize = 16)

plt.subplot(4, 2, 7)
grafico_vagas.plot.pie()
plt.title("Vagas em Pizza", fontsize = 16)

plt.subplot(4, 2, 8)
grafico_valor.plot.hist()
plt.title("Histograma de Valor", fontsize = 16)

plt.suptitle("GRÁFICOS", fontsize = 50)
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.9)

plt.show()
```
